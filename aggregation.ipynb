```python
import os
from os import listdir
from os.path import isfile,abspath, join,curdir
import sys
sys.path.insert(0, './common/')
import aggregation
import csv_utils
import pandas as pd
```

```python
FOLDER = os.path.join('data','submissions')
CURRENT_PATH = abspath(curdir)
FULL_PATH = join(CURRENT_PATH, FOLDER)
onlyfiles = [join(FULL_PATH, f) for f in listdir(FOLDER) if isfile(join(FULL_PATH, f))]
onlyfiles
```

```python
kaggleScores = [
    0.78603,# submission_0137PM-November-24-2018.csv
    0.78318,# submission_0230PM-November-25-2018.csv
    0.71005,# submission_0411PM-December-05-2018.csv
    0.64885,# submission_0603PM-November-23-2018.csv
    0.69281,# submission_0604PM-November-23-2018.csv
    0.66303,# submission_0843PM-November-23-2018.csv
    0.68705 # submission_0934PM-November-27-2018.csv
]
```

```python
df = pd.read_csv(onlyfiles[0])

files = onlyfiles[1:]
i = 1
for p in files:
    df = pd.merge(df, pd.read_csv(p), on='test_id', how='inner')
    i += 1
df.columns = ['test_id'] + ['col'+str(j) for j in range(i)]
    
df
```

```python
#aggregate = aggregation.aggregateAlbert
aggregate = aggregation.aggregateXavi

aggregated = aggregate(df, kaggleScores)
```
