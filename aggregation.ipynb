```python
import os
from os import listdir
from os.path import isfile,abspath, join,curdir
import sys
sys.path.insert(0, './common/')
import aggregation
import csv_utils
import pandas as pd
import numpy as np
```

```python
FOLDER = os.path.join('data', 'submissions_for_aggregation')
CURRENT_PATH = abspath(curdir)
FULL_PATH = join(CURRENT_PATH, FOLDER)
onlyfiles = [join(FULL_PATH, f) for f in listdir(FOLDER) if isfile(join(FULL_PATH, f))]
onlyfiles
```

```python
df = pd.read_csv(onlyfiles[0])

files = onlyfiles[1:]
i = 1
for p in files:
    df = pd.merge(df, pd.read_csv(p), on='test_id', how='inner')
    i += 1
df.columns = ['test_id'] + ['col'+str(j) for j in range(i)]
    
df.head(20)
```

```python
aggregate = aggregation.aggregateXavi
#aggregate = aggregation.aggregateAlbert

# Mean aggregation
# weights to each model (giving the same weight to all models)
scores = np.ones(7) # because there are 7 models
aggregated = aggregate(df, scores)
```

```python
csv_utils.create_csvs(aggregated['winner'], aggregated['test_id'])
```
