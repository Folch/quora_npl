```python
import pandas as pd

import os
import spacy
```

```python
path = os.path.join('..','data','train_data_v3.csv')
train_data = pd.read_csv(path)

path = os.path.join('..','data','test_data.csv')
test_data = pd.read_csv(path)
```

```python
!python -m spacy download en
```

```python
nlp = spacy.load('en', disable=['parser', 'ner', 'textcat'])
```

```python
#doc = nlp('How could I increase the speed of my internet')
doc = nlp(u"Oranges and I are similar. My boots and hippos aren't.")
' '.join([d.lemma_ for d in doc])
```

```python
for d in doc:
    print(d.lemma_, d.tag_)
```

```python
def nlp_func(question):
    doc = nlp(question)
    lemma = ' '.join([d.lemma_ for d in doc])
    tag = ' '.join([d.tag_ for d in doc])
    return pd.Series([lemma, tag])
```

```python
train_data[['question1_lemma', 'question1_tag']] = train_data.question1.apply(nlp_func)
```

```python
train_data[['question2_lemma', 'question2_tag']] = train_data.question2.apply(nlp_func)
```

```python
test_data[['question1_lemma', 'question1_tag']] = test_data.question1.apply(nlp_func)
```

```python
test_data[['question2_lemma', 'question2_tag']] = test_data.question2.apply(nlp_func)
```
